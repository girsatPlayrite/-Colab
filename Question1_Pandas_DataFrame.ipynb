```python
# @title Default title text
import pandas as pd

#1 Create a pandas DataFrame
data = {
    "StudentID": [101,102,103,104,105,106,107,108,109,110],
    "Name": ["Alice","Bob","Charlie","Diana","Ethan","Fiona","George","Hannah","Ivan","Julia"],
    "Age": [20,21,19,22,20,21,23,19,22,20],
    "Marks": [88,75,92,65,78,85,70,95,60,80],
    "Grade": ["A","B","A","C","B","A","B","A","C","B"]
}

skdata = pd.DataFrame(data)
```

```python
#2 Save the DataFrame as a CSV file named students.csv

skdata.to_csv("students.csv", index=False)
```

```python
#3 Reload students.csv into a new DataFrame
df = pd.read_csv("students.csv")
```

```python
#4 Display the first 3 rows using df.head(3)
print("First 3 rows:")
print(df.head(3))
```

```python
#5 Display the statistical summary using df.describe()
print("\nStatistical Summary:")
print(df.describe())
```

```python
#6 Explore the data further using df.shape, df.dtypes, and df.info()
print("\nDataFrame Shape:")
print(df.shape)

print("\nData Types:")
print(df.dtypes)

print("\nData Info:")
print(df.info())
```
