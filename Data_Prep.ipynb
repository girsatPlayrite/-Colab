```python
import pandas as pd

data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)
```

```python
#Task 1 — Categorical Encoding

#1. Gender → Label Encoding
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})

#2. City → One-Hot Encoding
city_dummy = pd.get_dummies(df['City'], prefix='City', drop_first=True)
df = pd.concat([df.drop('City', axis=1), city_dummy], axis=1)

#3. Size → Ordinal Encoding
size_mapping = {'Small': 0, 'Medium': 1, 'Large': 2}
df['Size'] = df['Size'].map(size_mapping)

print(df)
```

```python
#Task 2 — Feature Scaling

from sklearn.preprocessing import RobustScaler

#Initialize scaler
scaler = RobustScaler()

#Apply to Age and Fare
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

print(df)

#RobustScaler when outliers are present,.
#Fare contain extreme values(450).
```
