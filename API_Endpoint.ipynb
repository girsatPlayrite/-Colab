```python
#Task 1 — Fetch and Validate the Response

import requests

url = "https://restcountries.com/v3.1/all"

try:
    response = requests.get(url, timeout=10)

    # Print status code
    print(f"HTTP Status Code: {response.status_code}")

    if response.status_code == 200:
        data = response.json()

        # Validate that data is a list
        if isinstance(data, list):
            print(f"Total number of countries retrieved: {len(data)}")
        else:
            print("Unexpected response format: Expected a list of countries.")
    else:
        print(f"Request failed with status code {response.status_code}: {response.reason}")

except requests.exceptions.Timeout:
    print("Error: The request timed out.")
except requests.exceptions.ConnectionError:
    print("Error: Failed to establish a connection.")
except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")
```

```python
import requests
import pandas as pd

url = "https://restcountries.com/v3.1/all"

try:
    response = requests.get(url, timeout=10)
    print(f"HTTP Status Code: {response.status_code}")

    if response.status_code == 200:
        data = response.json()

        if isinstance(data, list):
            records = []

            # Extract required fields
            for country in data:
                common_name = country.get("name", {}).get("common", "Unknown")
                population = country.get("population", 0)
                region = country.get("region", "Unknown")
                area = country.get("area", 0)

                records.append({
                    "common_name": common_name,
                    "population": population,
                    "region": region,
                    "area": area
                })

            # Create DataFrame
            df = pd.DataFrame(records)

            # Display first 10 rows
            print("\nFirst 10 rows of the DataFrame:")
            print(df.head(10))

        else:
            print("Unexpected response format: Expected a list.")

    else:
        print(f"Request failed with status code {response.status_code}: {response.reason}")

except requests.exceptions.Timeout:
    print("Error: The request timed out.")
except requests.exceptions.ConnectionError:
    print("Error: Failed to establish a connection.")
except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")
```
