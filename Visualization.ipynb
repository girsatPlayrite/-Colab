```python
#Task 1 — Your First Matplotlib Chart

import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
tips = sns.load_dataset("tips")

# Group data: total tips per day
tips_per_day = tips.groupby("day")["tip"].sum()

# Create figure
plt.figure(figsize=(8, 5))

# Create bar chart
plt.bar(tips_per_day.index, tips_per_day.values)

# Add labels and title
plt.title("Total Tips Collected per Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Tip Amount")

# Display chart
plt.show()
```

```python
#Task 2 — Seaborn Histogram

import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
tips = sns.load_dataset("tips")

# Create histogram with KDE
sns.histplot(data=tips, x="tip", bins=20, kde=True)

# Add title
plt.title("Distribution of Tip Amounts")

# Show plot
plt.show()
```

```python
#Task 3 — Seaborn Scatter Plot

import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
tips = sns.load_dataset("tips")

# Create scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")

# Add title
plt.title("Total Bill vs Tip Amount")

# Show plot
plt.show()
```
